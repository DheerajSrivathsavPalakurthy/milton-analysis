```python
import re
```

```python
xml = open('pl/pl-normalized.xml').read()
```

```python
# Extract spans 
matches = re.finditer(r'(<milestone n="(.*?)" unit="sp"/>)', xml)
ml = list(matches)
spans = []
for i, match in enumerate(ml): 
    speaker = match.group(2)
    start = match.span()[1]
    try: 
        end = ml[i+1].span()[0]
    except: 
        end = len(xml)
    #print(match, speaker, start, end)
    spans.append([speaker, start, end])
```

```python
speakerDict = {}
for item in spans: 
    if item[0] in speakerDict: 
        speakerDict[item[0]].append((item[1], item[2]))
    else: 
        speakerDict[item[0]] = [(item[1], item[2])]
```

```python
speakerDict
```

```python
speechDict = {}
for speaker, spans in speakerDict.items(): 
    for span in spans:  
        #print(span)
        text = xml[span[0]:span[1]]
        if speaker in speechDict: 
            speechDict[speaker].append(text)
        else: 
            speechDict[speaker] = [text]
        
```

```python
def remove_tags(text):
    TAG_RE = re.compile(r'<[^>]+>')
    return TAG_RE.sub('', text)
```

```python
plainDict = {}
for speaker, texts in speechDict.items(): 
    alltext = '\n'.join(texts)
    allplaintext = remove_tags(alltext)
    speaker = speaker.replace(' ', '_')
    with open('pl/sp/' + speaker, 'w') as f: 
        f.write(allplaintext)
```
