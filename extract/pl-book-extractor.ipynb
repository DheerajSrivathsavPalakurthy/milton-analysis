```python
from bs4 import BeautifulSoup
import re
```

```python
xml = open('pl/pl-normalized.xml').read()
```

```python
soup = BeautifulSoup(xml, "lxml")
```

```python
rawbooks = soup.findAll(attrs={"type":"Book"})
```

```python
books = [book.text for book in rawbooks]
```

```python
for i, book in enumerate(books): 
    i = i + 1
    if i < 10: 
        i = '0' + str(i)
    else: 
        i = str(i)
    filename = "book" + i + ".txt" 
    print(filename)
    path = 'pl/bk/'
    with open(path + filename, 'w') as f: 
        f.write(book)
```
